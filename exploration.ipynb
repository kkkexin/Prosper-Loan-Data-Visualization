## Bivariate Exploration

> In this section, investigate relationships between pairs of variables in your
data. Make sure the variables that you cover here have been introduced in some
fashion in the previous section (univariate exploration).

```python
numeric_vars = ['ProsperScore','CreditScoreRangeLower','CreditScoreRangeUpper', 'CurrentDelinquencies','AmountDelinquent', 'DelinquenciesLast7Years','StatedMonthlyIncome','BorrowerRate','EstimatedEffectiveYield','LoanCurrentDaysDelinquent']

plt.figure(figsize = [8, 6])
sb.heatmap(df[numeric_vars].corr(), annot = True, fmt = '.3f',cmap = 'vlag_r', center = 0)
plt.title('Numeric Variables Heatmap')
plt.show()
```

热图观察可见，ProsperScore和BorrowerRate有较强负相关性；CreditScoreRange和ProsperScore，Current Delinquencies,BorrowerRate有一定负相关性。但是CurrentDelinquencies，AmountDelinquent,DelinquenciesLast7years却都和ProsperScore未表现出相关性，仅CurrentDelinquencies,DelinquenciesLast7Years和CreditScoreRange表现出了一定负相关性。

```python
# plot matrix: sample 500 diamonds so that plots are clearer and
# they render faster
samples = np.random.choice(df.shape[0], 500, replace = False)
df_samp = df.loc[samples,:]

g = sb.PairGrid(data = df_samp, vars = numeric_vars)
g = g.map_diag(plt.hist, bins = 20);
g.map_offdiag(plt.scatter)
```

散点图所得结论与热力图基本一致，不存在特别强的相关性。ProsperScore和BorrowerRate有较强负相关性。CreditScoreRange和ProsperScore，Current Delinquencies,BorrowerRate有一定负相关性。

```python
# plot matrix of numeric features against categorical features.
# can use a larger sample since there are fewer plots and they're simpler in nature.
categoric_vars = ['ProsperRating (Alpha)', 'EmploymentStatus','LoanStatus']
samples = np.random.choice(df.shape[0], 1000, replace = False)
df_samp = df.loc[samples,:]

def boxgrid(x, y, **kwargs):
    """ Quick hack for creating box plots with seaborn's PairGrid. """
    default_color = sb.color_palette()[0]
    sb.boxplot(x, y, color = default_color)
    plt.xticks(rotation=70)

plt.figure(figsize = [16, 24])
g = sb.PairGrid(data = df_samp, y_vars = ['ProsperRating (numeric)','BorrowerRate','Investors'], x_vars = categoric_vars,
                size = 6, aspect = 1)
g.map(boxgrid)
plt.show()
```

分类箱线图可以看出,BorrowerRate越低，信用评级越高。EmploymentStatus下Not Employed和Other的ProsperScore最低，BorrowerRate借款利率最高。出乎意料，Part-time工作者表现良好，ProsperRating中位数较其他类别最高，借款利率中位数也最低，同时分布较为集中。LoanStatus中Chargedoff类别相比Completed,Current借款利率更高，ProsperScore更低。DefaultPastDue时间越长，通常ProsperScore越低，借款利率越高。
评级为AA,A的investor中位数最多，评级为E,D,HR的investor中位数次之，而评级为B，C的investor中位数却最少。
